```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('company.csv')
data.head()
```

```python
print(data.shape)
data.describe() #2
```

```python
display(data) #3
```

```python
ageMean = data['age'].mean()
data['age'].fillna(ageMean, inplace=True)
display(data) #4
```

```python
incomeMedian = data['income'].median()
data['income'].fillna(incomeMedian, inplace=True)
display(data) #5
```

```python
ageMode = data['gender'].mode()
data['gender'].fillna(ageMode[0], inplace=True)

display(data) #6 
```

```python
sns.boxplot(x="gender", y="overall_score", data=data)
#7
```

```python
sns.violinplot(y="income", data=data)
#8
```

```python
sns.violinplot(x='gender', y="income", data=data)
#9
```

```python
scoreColumn = data['overall_score']
data = data[(scoreColumn > scoreColumn.mean() - 2 * scoreColumn.std()) & (scoreColumn < scoreColumn.mean() + 2 * scoreColumn.std())]
display(data)
#10
```

```python
minIncome, maxIncome = data.income.quantile([0.001,0.999])

data = data[(data.income > minIncome) & (data.income < maxIncome)]
display(data)

#11 
```

```python
sns.boxplot(x="gender", y="overall_score", data=data)
#12
```

```python
sns.violinplot(y="income", data=data)
#13
```

```python
sns.violinplot(x='gender', y="income", data=data)
#14
```
