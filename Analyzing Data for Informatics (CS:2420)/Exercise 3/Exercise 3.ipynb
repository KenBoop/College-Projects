```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
titanic = sns.load_dataset('titanic')
```

```python
titanic.head() #1
```

```python
titanic.describe() #1
```

```python
plt.title('HMS Titanic passengers by class', size=20)

sns.set(style="whitegrid", color_codes=True)

sns.countplot(x="class", color="g", data=titanic)

plt.show() 

#2
```

```python
plt.title('HMS Titanic passengers by class', size=20)

sns.set(style="whitegrid", color_codes=True)

sns.countplot(x="class",  hue="survived", color="g", data=titanic)

plt.show()

#3
```

```python
plt.title('Titanic passengers by class', size=20)

a = titanic[titanic.pclass == 1]["pclass"].count()
b = titanic[titanic.pclass == 2]["pclass"].count()
c = titanic[titanic.pclass == 3]["pclass"].count()

sns.set(style="whitegrid", color_codes=True)

labels = 'First', 'Second', 'Third'
sizes = [a, b, c]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')

patches, texts = plt.pie(sizes)
plt.legend(patches, labels, loc="lower right")


plt.axis('equal')

plt.show()

#4
```

```python
plt.title('Titanic passengers by class', size=20)

a = titanic[titanic.pclass == 1]["pclass"].count()
b = titanic[titanic.pclass == 2]["pclass"].count()
c = titanic[titanic.pclass == 3]["pclass"].count()

sns.set(style="whitegrid", color_codes=True)

labels = 'First', 'Second', 'Third'
sizes = [a, b, c]

tot = sum(sizes)
autopct = lambda x: "%d" % round(x*tot/100,2)
plt.pie(sizes, labels=labels, autopct=autopct)

plt.axis('equal')

plt.show()

#5
```

```python
plt.title('Titanic passengers by class', size=20)

explodeTuple = (0.1, 0.0)

a = titanic[titanic.alive == "yes"]["alive"].count()
b = titanic[titanic.alive == "no"]["alive"].count()

sns.set(style="whitegrid", color_codes=True)

labels = 'Survived', 'Did not survive'
sizes = [a, b]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', explode = explodeTuple)


plt.axis('equal')

plt.show()

#6
```

```python
plt.title('Fares paid by passengers of the Titanic by deck', size=20)

sns.stripplot(x="deck", y="fare", jitter=False, data=titanic)

plt.show()

#7
```

```python
plt.title('Fares paid by passengers of the Titanic by deck', size=20)

sns.stripplot(x="deck", y="fare", jitter=True, data=titanic)

plt.show()

#8
```

```python
plt.title('Fares paid by passengers of the Titanic by deck', size=20)

sns.stripplot(x="deck", y="fare", hue="sex", jitter=True, data=titanic)

plt.show()

#9
```
